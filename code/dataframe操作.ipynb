```python
user_item_ = df_click.groupby('user_id')['article_id'].agg(
    list).reset_index()
user_item_dict = dict(zip(user_item_['user_id'], user_item_['article_id']))
```

```python
user_item_dict
```

```python
import pickle
f = open('/home/xiaoguzai/数据/新闻推荐/user_data/sim/offline/itemcf_sim.pkl', 'rb')
item_sim = pickle.load(f)
f.close()
```

```python
def func_if_sum(x):
    user_id = x['user_id']
    article_id = x['article_id']

    interacted_items = user_item_dict[user_id]
    interacted_items = interacted_items[::-1]

    sim_sum = 0
    for loc, i in enumerate(interacted_items):
        try:
            sim_sum += item_sim[i][article_id] * (0.7**loc)
        except Exception as e:
            pass
    return sim_sum


def func_if_last(x):
    user_id = x['user_id']
    article_id = x['article_id']

    last_item = user_item_dict[user_id][-1]

    sim = 0
    try:
        sim = item_sim[last_item][article_id]
    except Exception as e:
        pass
    return sim
```

```python
from pandarallel import pandarallel
 
# Initialization
pandarallel.initialize()
# 用户历史点击物品与待预测物品相似度
df_feature['user_clicked_article_itemcf_sim_sum'] = df_feature[['user_id', 'article_id']].parallel_apply(func_if_sum, axis=1)
df_feature['user_last_click_article_itemcf_sim'] = df_feature[['user_id', 'article_id']].parallel_apply(func_if_last, axis=1)
```

```python
f = open('/home/xiaoguzai/数据/新闻推荐/user_data/sim/offline/binetwork_sim.pkl', 'rb')
binetwork_sim = pickle.load(f)
f.close()   
```

```python
def func_binetwork_sim_last(x):
    user_id = x['user_id']
    article_id = x['article_id']

    last_item = user_item_dict[user_id][-1]

    sim = 0
    try:
        sim = binetwork_sim[last_item][article_id]
    except Exception as e:
        pass
    return sim    
```

```python
df_feature['user_last_click_article_binetwork_sim'] = df_feature[[
    'user_id', 'article_id'
]].parallel_apply(func_binetwork_sim_last, axis=1)
```
